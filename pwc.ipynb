# Data Cleaning

```python
# Company ID and Company Name
ID = df1.groupby('Company ID',as_index=False)['Company Name'].value_counts()['Company ID']

Name = df1.groupby('Company Name',as_index=False)['Company ID'].value_counts()['Company Name']

df1 = df1[df1['Company ID'].apply(lambda x :x not in ID[ID.duplicated()].values)]

df1 = df1[df1['Company Name'].apply(lambda x :x not in Name[Name.duplicated()].values)]
```

```python
# Fiscal Year
df1 = df1[df1['Fiscal Year'].apply(lambda x : x in np.arange(1999,2022))]
```

```python
# SIC Code
df1 = df1[df1['SIC Code'].astype(str).str.isdecimal()]

df1 = df1[df1['SIC Code'].astype(str).apply(lambda x :len(x)==4)] 
```

```python
# Trading Currency 
df1 = df1[df1['Trading Currency'].apply(lambda x :x in ['USD','GBP'])]
```

```python
# SP / CDS / APD / ARD / ADA
df1.dropna(subset=['SP', 'CDS', 'APD', 'ARD', 'ADA'],inplace=True)

df1[['SP', 'CDS', 'APD', 'ARD', 'ADA']] = df1[['SP', 'CDS', 'APD', 'ARD', 'ADA']].astype('int64')
```

# Data Processing

```python
df1 = df1.melt(id_vars=['Company ID', 'Company Name', 'Fiscal Year', 'Industry', 'SIC Code',
       'Trading Currency'],value_vars=['SP', 'CDS', 'APD', 'ARD', 'ADA'],var_name='Metric Name',
        value_name='Value')
```
